**8. Splitting the data into Dependent and Independent Variables** 

```python
df
```

```python
X = df.iloc[:, 0:9]
X
```

```python
y = df.iloc[:, 9]
y
```

**9. Scaling Independent Variables**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
X
```

**10. Splitting the data into train and test data**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print("dimension of the input array", X_train.ndim)
print("dimension of the output array ",y_train.ndim)
```

```python
from sklearn.linear_model import LinearRegression
```

**11. Build the Model**

```python
model = LinearRegression()
```

**12. Train the Model**

```python
model.fit(X_train, y_train)
```

**13. Test the Model**

```python
y_model_train_pred = model.predict(X_train)
y_model_test_pred = model.predict(X_test) 
```

```python
y_model_train_pred
y_model_test_pred
```

**14. Measure the Performance using Metrics**

```python
Y_train_pred = model.predict(X_train)
Y_val_pred = model.predict(X_test)
Y_test_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import explained_variance_score

import numpy as np

print("Training R-squared: ",round(r2_score(y_train,Y_train_pred),2))
print("Training Explained Variation: ",round(explained_variance_score(y_train,Y_train_pred),2))

print('Training Mean Squared Error:', round(mean_squared_error(y_train,Y_train_pred), 2)) 
print("Training RMSE: ",round(np.sqrt(mean_squared_error(y_train,Y_train_pred)),2))
print("Training MAE: ",round(mean_absolute_error(y_train,Y_train_pred),2))
```

```python
print("Validation R-squared: ",round(r2_score(y_test,Y_val_pred),2))
print("Validation Explained Variation: ",round(explained_variance_score(y_test,Y_val_pred),2))

print('Validation Mean Squared Error:', round(mean_squared_error(y_train,Y_train_pred), 2)) 
print("Validation RMSE: ",round(np.sqrt(mean_squared_error(y_test,Y_val_pred)),2))
print("Validation MAE: ",round(mean_absolute_error(y_test,Y_val_pred),2))
```

```python
print("Test R-squared: ",round(r2_score(y_test,Y_test_pred),2))
print("Test Explained Variation: ",round(explained_variance_score(y_test,Y_test_pred),2))

print('Test Mean Squared Error:', round(mean_squared_error(y_test,Y_test_pred), 2)) 
print("Test RMSE: ",round(np.sqrt(mean_squared_error(y_test,Y_test_pred)),2))
print("Test MAE: ",round(mean_absolute_error(y_test,Y_test_pred),2))
```
